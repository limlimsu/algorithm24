## 알고리즘 예제 3.3

문제 정의 : 문자열 매칭(억지 기법)

알고리즘 설명 

1. 문자열 매칭 알고리즘에 입력 문자열 T와 패텬 P가 입력된다. 

2. T을 처음부터 끝까지 돌려서 P 위치를 찾는다.

3. P와 T의 부분 문자열을 비교하여 일치하는지 확인하여 일치할 경우 인덱스를 반환하고, 없으면 -1을 반환한다.  


손으로 푼 예제 :



코드 개요 : 억지 기법 문자열 매칭 알고리즘 (string_matching( T, P ))

입력 변수 : text, pattern

출력 변수 : string_matching(text, pattern )

```python
#알고리즘 코드 

def string_matching( T, P ):
    n = len(T)
    m = len(P)
    for i in range(n-m+1) :
        j = 0
        while j<m and P[j]==T[i+j] :
            j = j+1
        if j == m :
            return i
    return -1
```

```python
#테스트 코드 

text = 'HELLO WORLD'
pattern = 'EL'
print ( pattern, 'in', text, '-->', string_matching(text, pattern ))
pattern = 'WD'
print ( pattern, 'in', text, '-->', string_matching(text, pattern ))
```

복잡도 분석 : O(n^2)
