## 알고리즘 예제 3.4

문제 정의 : 최근접 쌍의 거리 

알고리즘 설명 

1. 거리계산 함수 설정한다. 

2. 모든 점 쌍을 비교하여 가장 가까운 두 점 사이의 거리를 찾는다.

3. 모든 점 쌍을 돌면서 두 점 사이의 거리 계산한다.

4. 더 작은 거리가 발견되면 해당 거리를 최근접 거리로 갱신한다.

5. 최근접 거리를 반환한다. 


손으로 푼 예제 :



코드 개요 : 거리계산 함수 (distance(p1, p2)), 최근접 거리 찾는 함수(closest_pair(p))

입력 변수 : p

출력 변수 : closest_pair(p)

```python
#알고리즘 코드

import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
    return mindist
```

```python
#테스트 코드 

p = [(1, 4), (12, 30), (40, 50), (5, 1), (12, 10), (3, 4)]
print("최근접 거리:", closest_pair(p))
```
