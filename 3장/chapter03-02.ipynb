## 알고리즘 예제 3.2 

문제 정의 : 순차 탐색 

알고리즘 설명 

1. 주어진 리스트를 처음부터 끝까지 돌린다.

2. 각 요소를 확인해서 찾는 key를 비교한다.

3. key를 찾으면 해당 요소 인덱스를 반환하고, 아닌 경우 -1을 반환한다.

손으로 푼 예제 : 



코드 개요 : 순차 탐색 함수(sequential_search(A, key))

입력 변수 : A, key

출력 변수 : result

```python
#알고리즘 코드

def sequential_search(A, key):
    for i in range(len(A)) :
      if A[i] == key :
        return i
    return -1
```

```python
#테스트 코드 

A = [1, 2, 3, 4, 5]
key = 2

result = sequential_search(A, key)

print("결과: ", result)
```

복잡도 분석 : O(n)
