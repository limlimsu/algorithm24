## 알고리즘 예제 3.1

문제 정의 : 선택 정렬 

알고리즘 설명 

1. 리스트에서 가장 작은 값을 찾아 첫 번째 위치와 교환한다.

2. 그 다음으로 작은 값을 찾아 두 번째 위치와 교환한다. 

3. 1, 2번 과정을 마지막 값이 정렬될 때까지 반복한다. 

손으로 푼 예제 :




코드 개요 : 선택 정렬 함수 (selection_sort(A))

입력 변수 : A 

출력 변수 : data, selection_sort(data)

```python
#알고리즘 코드

def selection_sort(A) :
    n = len(A)
    for i in range(n-1) :
        least = i
        for j in range(i+1, n) :
            if (A[j]<A[least]) :
                least = j
        A[i], A[least] = A[least] , A[i]
        printStep(A, i + 1)
        
def printStep(arr, val) :
    print(" Step %2d = " % val, end='')
    print(arr)       
```

```python
#테스트 코드 
    
data = [ 4, 2, 5, 3, 1, 7, 8, 9, 6 ]
print("Original : ", data)
selection_sort(data)
print("Selection : ", data)
```
