## 알고리즘 예제 3.6

문제 정의 : 깊이 우선 탐색

알고리즘 설명 

1. 시작 노드를 방문하고 이웃한 노드 중 아직 미방문인 노드를 재귀적으로 방문한다.

2. 모든 이웃 방문 후 이전 노드로 돌아가서 그 노드의 다른 이웃을 방문한다.

3. 1, 2번 과정을 반복하여 그래프의 노드를 방문한다. 

손으로 푼 예제 : 



코드 개요 : 깊이 우선 탐색 함수 (dfs( graph, start, visited ))

입력 변수 : graph, start, visited

출력 변수 : 

```python
#알고리즘 코드

def dfs( graph, start, visited ):
    if start not in visited :
        visited.add(start)
        print(start, end=' ')
        nbr = graph[start] - visited
        for v in nbr:
            dfs(graph, v, visited)
```

```python
#테스트 코드 

mygraph = { "A" : {"B", "C"}, 
            "B" : {"A", "D"},
            "C" : {"A", "D", "E"},  
            "D" : {"B", "C", "F"},
            "E" : {"C", "G", "H"},
            "F" : {"D"},
            "G" : {"E", "H"},
            "H" : {"E", "G"},
           }     

print('DFS : ', end='')
dfs(mygraph, "A", set() )
print()
```
