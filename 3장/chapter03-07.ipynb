## 알고리즘 예제 3.7

문제 정의 : 너비 우선 탐색

알고리즘 설명 

1. 시작 노드에서 모든 인접한 노드를 탐색 후 그 인접한 노드들의 인접한 노드를 탐색한다.

2. 1번의 방식으로 시작 노드로부터 일정 거리에 있는 노드 모두 탐색 후 시작 노드로부터 더 멀리 떨어진 노드들을 탐색한다.

+ BFS는 큐(Queue)를 사용하여 구현된다. 

손으로 푼 예제 : 



코드 개요 : 너비 우선 탐색 함수 (bfs( graph, start))

입력 변수 : graph, start

출력 변수 : 

```python
#알고리즘 코드

import queue
def bfs(graph, start):
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty():
        v = que.get()
        print(v, end=' ')
        nbr = graph[v] - visited
        for u in nbr:
            visited.add(u)
            que.put(u)
```

```python
#테스트 코드 

mygraph = { "A" : {"B", "C"}, 
            "B" : {"A", "D"},
            "C" : {"A", "D", "E"},  
            "D" : {"B", "C", "F"},
            "E" : {"C", "G", "H"},
            "F" : {"D"},
            "G" : {"E", "H"},
            "H" : {"E", "G"},
           }    
print('BFS : ', end='')
bfs(mygraph, "A")
print() 
```
