## 알고리즘 예제 6.2

문제 정의 : 카운팅 정렬 

알고리즘 설명 

1. count의 크기는 MAX_VAL로 하고, 모두 0으로 초기화한다. 

2. 각 숫자의 빈도를 저장 

3. for i in A: count[i] += 1로 각 숫자별 빈도를 계산한다.

4. for i in range(MAX_VAL): count[i] += count[i-1]로 count[i]가 출력 배열에서 해당 숫자의 위치가 되도록 수정한다. 

5. for i in range(len(A)):  output[count[A[i]]-1] = A[i]  count[A[i]] -= 1로 모든 입력항목 A[i]에 대해 해당 위치(count[A[i]]-1)에 저장하고, 킷값 A[i]의 위치를 하나 줄인다.

6. 마지막으로 정렬 결과를 원래 배열에 복사한다.

손으로 푼 예제 :





코드 개요 : 카운팅 정렬 함수 (counting_sort(A))

입력 변수 : 최대 빈도수(MAX_VAL = 10), data = [ 7, 8, 8, 5, 5, 5, 1, 3, 3]

출력 변수 : 기존 data, 카운팅 정렬된 data(counting_sort(data))

```python
#알고리즘 코드

def counting_sort(A):
    output = [0] * len(A)
    count = [0] * MAX_VAL

    for i in A:
        count[i] += 1

    for i in range(MAX_VAL):
       count[i] += count[i-1]

    for i in range(len(A)):
        output[count[A[i]]-1] = A[i]
        count[A[i]] -= 1

    for i in range(len(A)):
        A[i] = output[i] 
```

```python
#테스트 코드

MAX_VAL = 10
data = [ 7, 8, 8, 5, 5, 5, 1, 3, 3]
print("Original :", data)
counting_sort(data)
print("Counting :", data)
```

복잡도 분석 

카운팅 정렬은 입력 데이터가 크지 않은 일정한 범위의 값을 가진 정수라면 매우 효율적이다. 
알고리즘의 모든 루프가 단일루프이다. 만약 입력의 크기가 n이고, 숫자의 범위가 k가지(0~ k-1)라면 전체 알고리즘은 O(k+n)의 시간 복잡도를 갖는다.
따라서 아스키 문자의 정렬에서와 같이 K가 작은 수로 제한된다면 카운팅 정렬 함수의 시간 복잡도는 O(n)이 된다.
