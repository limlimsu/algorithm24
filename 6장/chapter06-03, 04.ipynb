## 알고리즘 예제 6.3, 4

문제 정의 : 호스풀 알고리즘의 시프트 테이블 만들기, 호스풀 알고리즘 

알고리즘 설명 

시프트 테이블 

1. 크기가 NO_OF_CHARS인 리스트를 만들고, 모든 항목을 m으로 채운다. 이것은 일단 모든 문자는 패턴에 없다고 생각하는 것이다. 

2. 패턴의 모든 문자에 대해 테이블의 해당 위치의 이동 거리를 계산해 넣는다. 만약 패턴에 중복된 문자가 있다면 맨 오른쪽 문자에 의한 거리가 최종적으로 들어간다.(반복문이 0부터 증가하는 방향으로 진행되므로)

호스풀 알고리즘

1. 패턴과 텍스트의 길이를 구하고, 시프트 테이블을 계산한다.

2. 패턴의 위치는 우측 끝 문자를 기준으로 한다

3. 패턴을 뒤에서 아픙로 맞지 않을 때까지 검사한다.

4. 패턴의 모든 문자가 일치하면 위치를 반환하고, 일치하지 않으면 테이블을 찾아 그 만큼 건너뛴다. 

손으로 푼 예제 :







코드 개요 : 시프트 테이블(def shift_table(pat)), 호스풀 함수(def search_horspool(T, P))

입력 변수 : 텍스트(m = len(P)), 패턴(n = len(T))

출력 변수 : search_horspool("PEACHPINEAPPLEGRAPE", "GRAPE")

```python
#알고리즘 코드

NO_OR_CHARS = 128
def shift_table(pat):
    m = len(pat)
    tbl = [m]*NO_OR_CHARS
    
    for i in range(m-1):
        tbl[ord(pat[i])] = m-1-i

    return tbl

def search_horspool(T, P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1
    while(i <= n-1):
        k = 0
        while k <= m-1 and P[m-1-k]==T[i-k]:
            k += 1
        if k == m :
            return i-m+1
        else :
            tc = t[ord(T[i-k])]
            i += max(1, (tc-k))
    return -1
```

```python
#테스트 코드

print("패턴의 위치", search_horspool("PEACHPINEAPPLEGRAPE", "GRAPE"))
```
