## 알고리즘 예제 5.3

문제 정의 : 퀵 정렬

알고리즘 설명 

1. 정렬할 항목이 2개 이상인 경우에 대해서만 처리한다. 항목이 1개이면 이미 정렬된 것이다.

2. 분할 알고리즘을 이용해 리스트를 좌우로 분할한다. 이때 리스트가 균등하게 분할된다는 보장은 없다. 분할이 완료되면 피벗의 인덱스가 반환되어 mid에 복사된다. 

3. 왼쪽 부분 리스트 A[left..mid-1]와 오른쪽 부분 리스트 A[mid+1..right]를 각각 정렬한다. 

손으로 푼 예제 :



코드 개요 : 퀵 정렬 함수 (quick_sort(A, left, right)), 분할 함수 (partition(A, left, right))

입력 변수 : data = [ 1, 3, 4, 2, 5, 10, 8, 9, 7, 6 ]

출력 변수 : 기존 data, 퀵 정렬(quick_sort(data, 0, len(data)-1))된 data

```python
#알고리즘 코드

def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]
    return high

def quick_sort(A, left, right) :
    if left<right :
        mid = partition(A, left, right) 
        quick_sort(A, left, mid-1)
        quick_sort(A, mid+1, right)
```

```python
#테스트 코드

data = [ 1, 3, 4, 2, 5, 10, 8, 9, 7, 6 ]
print("Original : ", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort :", data)
```

최선의 경우: 피벗을 항상 중간 값으로 선택하고 분할할 때마다 거의 같은 크기의 하위 배열을 생성하는 경우로 이 때 퀵 정렬은 O(nlog2(n))의 성능을 보인다.

평균적인 경우: 피벗이 어느 정도 무작위로 선택되고 배열이 균등하게 분할되는 경우로 이 때 퀵 정렬은 O(nlog2(n))의 성능을 보인다.

최악의 경우: 피벗이 항상 배열의 최솟값이나 최댓값으로 선택되어 배열이 한쪽으로 치우쳐진 경우로 이 때 퀵 정렬은 O(n^2)의 성능을 보인다.
