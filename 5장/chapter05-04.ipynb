## 알고리즘 예제 5.4

문제 정의 : 이진트리의 높이를 구하는 함수 

알고리즘 설명 

1. 공백 트리이면 0을 반환한다. (높이가 0이라는 뜻)

2. 왼쪽 서브 트리의 높이(hLeft)와 오른쪽 서브 트리의 높이(hRight)를 구한 후 최댓값에 1을 더해서 반환한다. 

손으로 푼 예제 :



코드 개요 : 이진트리 노드 클래스 (class TNode), 이진트리 높이 구하는 함수(calc_height(root))

입력 변수 : 
이진트리 
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c) 

출력 변수 : calc_height(root)

```python
#알고리즘 코드

class TNode:
    def __init__ (self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
```

```python
#테스트 코드

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)
print(" 트리의 높이 =", calc_height(root))
```
