## 알고리즘 예제 8.1

문제 정의 : 거스름돈 최소화 알고리즘(탐욕적 기법)

알고리즘 설명 

1. coins는 모든 동전의 액면가 리스트로 반드시 내림차순으로 정렬되어 있어야 한다. V는 거스름돈의 액수이다.

2. 각 액면가 동전의 개수를 저장하는 리스트를 만든다.

3. 액멱가가 높은 동전부터 모든 동전에 대해 반복을 한다. 

4. 현재 액멱가를 최대한 사용하는 동전 개수를 계산해 리스트에 넣고, 남은 거스름돈을 계산한다.

5. 마지막으로 각 액면가 동전의 개수 리스트를 반환한다. 

손으로 푼 예제 :



코드 개요 : 탐욕적 기법의 거스름돈 최소화 알고리즘(min_coins_greedy( coins, V ))

입력 변수 : coins = [500, 100, 50, 10, 5, 1] (모든 동전 액면가 리스트), V (거스름돈의 액수)

출력 변수 : min_coins_greedy( coins, V ) (거스름돈 최소화 알고리즘)

```python
#알고리즘 코드

def min_coins_greedy( coins, V ):
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count
```

```python
#테스트 코드

coins = [500, 100, 50, 10, 5, 1]
V = 570
print("잔돈 액수 = ", V)
print("동전 종류 = ", coins)
print("동전 개수 = ", min_coins_greedy(coins, V ))
```
