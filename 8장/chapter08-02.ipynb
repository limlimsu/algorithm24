## 알고리즘 예제 8.2

문제 정의 : 분할 가능한 배낭 채우기(탐욕적 기법)

알고리즘 설명 

1. coins는 모든 동전의 액면가 리스트로 반드시 내림차순으로 정렬되어 있어야 한다. V는 거스름돈의 액수이다.

2. 각 액면가 동전의 개수를 저장하는 리스트를 만든다.

3. 액멱가가 높은 동전부터 모든 동전에 대해 반복을 한다. 

4. 현재 액멱가를 최대한 사용하는 동전 개수를 계산해 리스트에 넣고, 남은 거스름돈을 계산한다.

5. 마지막으로 각 액면가 동전의 개수 리스트를 반환한다. 

손으로 푼 예제 :



코드 개요 : 탐욕적 기법의 분할 가능한 배낭 채우기 알고리즘(knapSack_fractional_greedy(obj, W))

입력 변수 : 파이썬 튜플 형태의 obj (물건, 무게, 가치)

출력 변수 : 부분적인 배낭(18, 50) (knapSack_fractional_greedy(obj, 18, 50))

```python
#알고리즘 코드

def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)

    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
           W -= o[1]
           totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))

    return totalValue
```

```python
#테스트 코드

obj = [ ("A", 10, 80), ("B", 12, 120), ("C", 8, 60)] # (물건,무게,가치)
print("W = 18", obj)
print("부분적인배낭(18): ", knapSack_fractional_greedy(obj, 18), end='\n\n')

obj = [ ("A", 10, 60), ("B", 40, 40), ("C", 20, 100), ("D", 30, 120)] 
print("W = 50", obj)
print("부분적인배낭(50): ", knapSack_fractional_greedy(obj, 50))
```

복잡도 분석: 

n개의 물건이 주어지는 경우, 먼저 obj.sort(key = lambda o: o[2]/o[1], reverse=True)에서 정렬이 필요하므로 O(nlog2(n))시간이 필요하다. 다음 과정은 n번 반복하는 for o in obj :이므로 O(n)이다. 결국, 전체 알고리즘의 시작 복잡도는 정렬 방법에 달려 있고, 따라서 O(nlog2(n))이 된다.
