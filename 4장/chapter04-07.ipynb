## 알고리즘 예제4.7

문제 정의 : 거듭제곱(억지 기법)

알고리즘 설명

1. result를 1로 초기화한다.

2. 전체 범위에서 n번 반복하여 결과값 반환한다. 

손으로 푼 예제 : 



코드 개요 : 거듭제곱 함수(slow_power(x, n))

입력 변수 : t2, t3

출력 변수 : t3-t2, slow_power(2.0, 500)

```python
#알고리즘 코드

def slow_power(x, n) :
    result = 1.0
    for i in range(n):
        result = result * x
    return result
```

```python
#테스트 코드

import time
print("축소정복기법(2**500) =", slow_power(2.0, 500))

t2 = time.time()
for i in range(100000) : slow_power(2.0, 500)
t3 = time.time()

print("   억지기법 시간...", t3-t2)
```

복잡도 분석

반복문에 의해 n에 비례하는 시간이 걸리고, 따라서 시간 복잡도가 O(n)이다.
