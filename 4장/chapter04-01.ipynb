## 알고리즘 예제 4.1

문제 정의 : 팩토리얼(하향식 축소 기법)

알고리즘 설명 

1. n이 1이면 1로 반환하고 1이 아니면 n * factorial_recur(n - 1)을 계산하여 반환한다. 

손으로 푼 예제 :



코드 개요 : 하향식 축소 정복 기법 함수 (factorial_recur)

입력 변수 : n

출력 변수 : n * factorial_recur(n - 1)

```python
#알고리즘 코드

def factorial_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n * factorial_recur(n - 1))
```

```python
#테스트 코드 

n = 5
print(n, "factoriaal is ", factorial_recur(n))
```

수행 결과



복잡도 분석

n*(n-1)*(n-2)*...1 이므로 복잡도는 O(n!)이 된다.
