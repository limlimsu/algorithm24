## 알고리즘 예제4.6

문제 정의 : 이진 탐색(반복구조)

알고리즘 설명

1. binary_search(A, key, low, high)함수에는 정렬된 리스트 A와 킷값 key, 그리고 리스트 내의 탐색 범위인 low와 high가 입력된다. 

2. low와 high가 역전되지 않으면, 아직 탐색할 영역이 남은 경우이다.

3. mid를 계산해 그 항목(A[mid])이 key와 같으면 탐색 성공이다.

4. key가 A[mid]보다 크면 mid+1 ~ high사이를 검색하고, key가 A[mid]보다 작으면 low ~ mid-1을 검색한다. 

5. 모든 탐색이 실패하면 -1을 반환한다. 

손으로 푼 예제 : 



코드 개요 : 이진탐색 함수(반복구조) ()

입력 변수 : n

출력 변수 : factorial_iter(n)

```python
#알고리즘 코드

def binary_search_iter(A, key, low, high) :
    while (low <= high) :
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1 
```

```python
#테스트 코드

listA = [1, 4, 7, 12, 15, 15, 25, 28, 30, 33, 40, 47, 50, 77]
print("입력 리스트 =", listA)
print("7 탐색(반복) -->", binary_search_iter(listA, 7, 0, len(listA)-1) )
print("3 탐색(반복) -->", binary_search_iter(listA, 3, 0, len(listA)-1) )
```
