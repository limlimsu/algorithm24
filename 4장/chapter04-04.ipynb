## 알고리즘 예제4.4

문제 정의 : 위상 정렬(축소 정복 기법)

알고리즘 설명

1. inDeg 딕셔너리를 초기화한다.

2. 그래프의 모든 간선을 이용해 정점의 진입차수를 계산한다.

3. inDeg에서 진입차수가 0인 모든 정점을 찾아 vlist를 만든다.

4. vlist가 공백이 아니면 정점 v를 꺼내고 화면에 출력한다.

5. v의 모든 인접 정점 u에 대해 진입차수를 감소시키고 u의 진입차수가 0으로 줄어들면 vlist에 u를 추가한다. 

손으로 푼 예제 : 




코드 개요 : 위상 정렬함수(축소 정복 기법) (topological_sort(graph))

입력 변수 : mygraph

출력 변수 : topological_sort(mygraph)

```python
#알고리즘 코드

def topological_sort(graph) :
    inDeg = {}
    for v in graph :
        inDeg[v] = 0
    for v in graph :
        for u in graph[v]:
            inDeg[u] += 1
            
    vlist = []
    for v in graph :
        if inDeg[v] == 0 :
            vlist.append(v)
            
    while vlist :
        v = vlist.pop()
        print(v, end='')
        
        for u in graph[v] :
            inDeg[u] -= 1
            if inDeg[u] == 0 :
               vlist.append(u)
```

```python
#테스트 코드

mygraph = { "A" : {"C","D"},
            "B" : {"D","E"},
            "C" : {"D","F"},
            "D" : {"F"},
            "E" : {"F"},
            "F" : {}
}
print('topological_sort: ')
topological_sort(mygraph)
print() 
```

복잡도 분석

모든 정점 초기화가 n번 반복되고, for v in graph : for u in graph[v]:
inDeg[u] += 1의 이중 루프는 모든 간선에 대해 처리하므로 간선의 개수만큼 반복되어서 전체 복잡도는 O(n+e)가 된다.
