## 알고리즘 예제4.8

문제 정의 : 거듭제곱(축소 정복 기법)

알고리즘 설명

1. n이 0이면 1을 반환한다. 

2. n이 짝수이면 x의 제곱을 재귀적으로 계산한다. 

3. n이 홀수이면 x와 제곱수를 계산하여 재귀적으로 호출한다. 

손으로 푼 예제 : 



코드 개요 : 거듭제곱(축소 정복 기법)함수 (power(x, n))

입력 변수 : t1, t2

출력 변수 : t2-t1, power(2.0, 500)

```python
#알고리즘 코드

def power(x, n) :
    if n == 0 :
        return 1
    elif (n % 2) == 0 :
        return power(x*x, n//2)
    else :
        return x * power(x*x, (n-1)//2)
```

```python
#테스트 코드

import time
print("   억지기법(2**500) =", power(2.0, 500))

t1 = time.time()
for i in range(100000) : power(2.0, 500)
t2 = time.time()

print("축소정복기법 시간", t2-t1)
```

복잡도 분석

n을 2의 거듭제곱 값의 하나인 2^k라고 가정해 보자. 순환호출을 한번 할 때마다 n의 크기가 절반씩 줄어들게 되므로 2^k -> 2^(k-1) -> 2^(k-2) -> ... -> 2^0으로 약 k번의 호출이 일어나게 됨을 알 수 있다. n = 2^k이므로 양변에 log를 취해주면 log2(n) = k가 된다. 

따라서 복잡도는 O(log2(n))이 된다.
