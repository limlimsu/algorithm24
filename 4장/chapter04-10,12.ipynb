## 알고리즘 예제4.10, 12

문제 정의 : 정렬을 이용한 K번째 작은 수 찾기 

알고리즘 설명

10번 

1. 입력으로 주어진 리스트 A를 정렬한다. 

2. 정렬된 리스트에서 K-1 인덱스에 해당하는 요소를 반환한다. (K-1을 하는 이유는 인덱스가 0부터 시작하기 때문이다.)

12번 

1. low, high와 피벗의 위치 pivot을 설정한다. 피벗은 첫 번째 항목으로 선택하였다. 

2. low 와 high가 역전되지 않는 동안 항목의 크기가 피벗을 기준으로 반대인 위치까지 이동한다. low는 증가, high는 감소 방향으로 탐색한다. 

3. 선택된 두 레코드를 교환한다.

4. 최종적으로 high와 피벗의 항목을 교환하고, high(피벗의 인덱스)를 반환한다.

손으로 푼 예제 : 



코드 개요 : 정렬 이용한 k번째 작은 수 찾기 함수 (kth_smallest_sort(A, k))

입력 변수 : array = [11, 5 ,7, 8, 19, 17, 25, 20]

출력 변수 : kth_smallest_sort(array, 2), kth_smallest_sort(array, 4)

```python
#알고리즘 코드

def kth_smallest_sort(A, k):
    A.sort()
    return A[k-1]

def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]
    return high
```

```python
#테스트 코드

array = [11, 5 ,7, 8, 19, 17, 25, 20]
print("입력 리스트 =", array)
print("[정렬기법] 2번째 작은 수: ", kth_smallest_sort(array, 2))
print("[정렬기법] 4번째 작은 수: ", kth_smallest_sort(array, 4))
```
