## 알고리즘 예제4.3

문제 정의 : 삽입 정렬 

알고리즘 설명

1. 주어진 리스트 A의 길이를 n이라고 한다.

2. 리스트 A의 두 번째 원소부터 마지막 원소까지를 차례대로 선택한다. 

3. 선택한 원소를 key 변수에 저장한다. 

4. key 변수와 key보다 작은 원소들을 비교하여 key보다 큰 원소를 오른쪽으로 한 칸씩 이동한다. 

5. key보다 작은 원소를 찾거나 리스트의 시작에 도달하면 그 위치에 key를 삽입한다. 

6. 위의 과정을 반복하여 리스트 전체를 정렬한다.

손으로 푼 예제 : 



코드 개요 : 삽입 정렬 함수 (insertion_sort(A))

입력 변수 : data

출력 변수 : data, insertion_sort(data) 

```python
#알고리즘 코드

def insertion_sort(A) :
    
    n = len(A)
    for i in range(1, n) :
        key = A[i]
        j = i - 1
        while j>=0 and A[j] > key :
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key
        printStep(A, i)             
```
