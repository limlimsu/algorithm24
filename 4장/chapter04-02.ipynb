## 알고리즘 예제4.2

문제 정의 : 팩토리얼(상향식 축소 기법)

알고리즘 설명

1. result 변수를 1로 초기화한다

2. 1~n 까지 숫자에 대해 반복문 실행한다

3. 각 숫자를 result를 곱하여 결과를 갱신한다.

4. 반복문이 끝나면 최종 결과인 result를 반환한다. 

손으로 푼 예제 : 



코드 개요 : 상향식 축소 정복 기법 함수 (factorial_iter(n))

입력 변수 : n

출력 변수 : factorial_iter(n)

```python
#알고리즘 코드

def factorial_iter(n) :
    result = 1
    for k in range(1,n+1):
        result = result * k
    return result
```

```python
#테스트 코드

n = 8
print(n, "factoriaal is ", factorial_iter(n))  
```

수행 결과 



복잡도 분석 

1~n까지의 숫자를 반복하며 각 숫자를 곱하므로 O(n!)이 된다.
