## 알고리즘 예제 7.10

1. 문제정의: 

- Floyd 최단경로 알고리즘 

2. 알코리즘 성명:

- 정점 리스트 `vertex` 와 인접행렬 `W`
- `for k in range(vsize):` 모든 정점에 대해 
- `for i in range(vsize):`, `for j in range(vsize):` 최단거리 행렬의 모든 행 갱신

3. 손으로 푼 예제 (이미지삽입):




4. 코드 개요(함수, 입력변수, 출력변수):


- 함수 : 
    - `def shortest_path_floyd(vertex, W):` 모든 정점 쌍 사이의 최단 경로를 계산

- 입력 변수 :  
    - `vertex` : `vertex =  [ 'A', 'B', 'C', 'D', 'E', 'F', 'G']` 정점의 리스트 각 노드 
    - `W` : 그래프의 가중치 행렬, 각 정점 쌍 사이의 가중치
            
            
        
- 출력 변수 : 
    - `printD(D)` 정점 쌍 사이의 최단 경로를 저장한 거리 행렬 출력


```python
# 5. 코드
import copy

def shortest_path_floyd(vertex, W):
    vsize = len(vertex)
    D = copy.deepcopy(W)

    for k in range(vsize):
        for i in range(vsize):
            for j in range(vsize):
                if (D[i][k] + D[k][j] < D[i][j]):
                    D[i][j] = D[i][k] + D[k][j]
    printD(D)
```

```python
# 6. 테스트 코드 

def printD(D):
    vsize = len(D)
    print("==================================")
    for i in range(vsize):
        for j in range(vsize):
            if (D[i][j] == INF):
                print("INF ", end='')
            else:
                print("%4d" % D[i][j], end='')
        print("")

INF = 9999
vertex =  [ 'A', 'B', 'C', 'D', 'E', 'F', 'G']
weight =  [ [ 0, 7, INF, 3, 10, INF, INF ],
            [ 7, 0, 4, 2, 6, INF, INF],
            [INF, 4, 0, 11, INF, INF, INF],
            [3, 2, 11, 0, 13, 4, 5],
            [10, 6, INF, 13, 0, 9, INF],
            [INF, INF, INF, 4, 9, 0, 2],
            [INF, INF, INF, 5, INF, 2, 0] ]

print("Shortest Path By Floyd's Algorithm")
shortest_path_floyd(vertex, weight)
```
