## 알고리즘 예제 7.5


1. 문제정의: 

0-1 배낭 채우기 알고리즘 (분할 정복 기법)


2. 알코리즘 성명:
- `if n == 0 or W == 0 :` 기반 상황으로 물건이 없거나 배낭 용량이 0이면 `return 0` 최대가치는 0이다. 
- n번째 항목이 배낭 용량보다 크면 항목을 배낭에 넣을 수 없고 남은 n-1개의 항목들로 다시 계산
- 그렇지 않으면 이 항목을 넣지 않는 경우의 배낭 가치 와 넣는 경우의 가치를 계산하고 이들 중에서 큰값을 선택하고 반환한다. 

3. 손으로 푼 예제 (이미지삽입):



4. 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def knapSack_dc(W, wt, val, n):` 

- 입력 변수 :  
    - `W` 현재 배낭의 용량
    - `wt` 물건들의 무게를 저장한 리스트
    - `val` 물건들의 가치를 저장한 리스트 
    - `n` 물건들의 수 
- 출력 변수 : 
    - `max(valwith, valWithout)` 항목을 넣지 않는 경우와 넣는 경우의 가치들 중에서 큰 값


```python
def knapSack_dc(W, wt, val, n):
    if n == 0 or W == 0 :
        return 0 # 종료 조건 
    
    if (wt[n-1] > W):
        return knapSack_dc(W, wt, val, n-1)
    else:
        valWithout = knapSack_dc(W, wt, val, n-1)
        valwith= val[n-1] + knapSack_dc(W-wt[n-1], wt ,val, n-1)
        return max(valwith, valWithout)
    
```

```python
# 6. 테스트 코드 
import random
import time

random.seed(0)  

val = [random.randint(1, 100) for _ in range(50)] # 물건들의 가치 리스트 
wt = [random.randint(1, 10) for _ in range(50)] # 물건들의 무게 리스트
W = 25 # 현재 배낭의 용량 
n = len(val)

start = time.time()
result = knapSack_dc(W, wt, val, n)
end = time.time()

print("0-1 배낭문제(분할정복):", result)
print("실행 시간: %.5f 초" % (end - start))
```

8. 복잡도 분석:

재귀함수로 항목을 넣는 경우와 넣지 않는 경우가 두 가지 이므로 2^n이다. 그러기 때문에 시간 복잡도는 O(2^n)이 된다.
