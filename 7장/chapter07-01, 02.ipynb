## 알고리즘 예제 7.1, 2

문제 정의 : 피보나치수열(메모이제이션 이용), 피보나치수열(테이블화 이용)

알고리즘 설명 

메모이제이션 함수 

1. 아직 풀리지 않은 문제인지 검사한다. mem은 크기가 n+1인 리스트로, 전역변수로 선언되어 있어야 한다. mem[k]는 k번째 피보나치 수를 의미한다.

2. 아직 풀리지 않았다면 풀어야 한다. 먼저 순환식의 초기 조건(기반 상황)을 검사해 해당하면 결과를 바로 계산한다. 일반 상황이면 순환 호출한다.

3. 이미 풀린 경우이거나, 처음 푼 경우 모두 결과는 mem[n]에 저장되어 있다. 이를 반환한다. 

테이블화 함수 

1. n+1 크기의 리스트(테이블)를 만들고 일단 모든 항목들을 None으로 초기화한다. 다음으로 기반 상황에 대한 답을 저장한다.

2. 상향식으로 답을 테이블에 저장해나간다. for i in range(2, n+1):에서 i가 2, 3, 4 ...로 증가해 나간다. 

손으로 푼 예제 :





코드 개요 : 메모제이션 함수(fib_dp_mem(n)), 테이블화 함수(fib_dp_tab(n))

입력 변수 : n = 7, 

출력 변수 : fib_dp_tab(n), fib_dp_mem(n)

```python
#알고리즘 코드

def fib_dp_mem(n) :
    if( mem[n] == None ) :
        if n < 2 :
            mem[n] = n
        else :
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]

def fib_dp_tab(n) :
    f = [None] * (n+1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n+1):
        f[i] = f[i-1] + f[i-2]
    return f[n]
```

```python
#테스트 코드

n = 7
print('동적계획(  테이블화  ): Fibonacci(%d) = '%n, fib_dp_tab(n))
mem = [None] * (n+1)
print('동적계획(메모이제이션): Fibonacci(%d) = '%n, fib_dp_mem(n))
```

복잡도 분석

메모이제이션: 모든 부분 문제를 정확히 한 번씩 처리하고, 이러한 부분 문제의 수가 n개 이므로 O(n)이다.

테이블화: 순환호출을 사용하지 않고, n번 반복하는 하나의 루프(for i in range(2, n+1):)이므로 O(n)이다.
