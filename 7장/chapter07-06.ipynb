## 알고리즘 예제 7.6

1. 문제정의: 

0-1 배낭 채우기 알고리즘 (동적 계획법)

2. 알코리즘 성명:
- (n+1)(W+1) 크기의 2차원 리스트를 생성 
- i번쨰 물건이 배낭 용량을 초과하면, 이 물건을 제외한 남은 물건을 고려한 경우의 최대 가치를 구한다. 그런데 이 값은 이미 계산되어 있다.
- 만약 그렇지 않다면 이 항목을 넣는 경우의 가치합 과 넣지 않는 경우의 가치합을 계산한다. 이것도 계산되어 있어 표에서 참조
- `A[i][w] = max(valWith, valWithout)` 이들중에서 큰값을 선택하여 `A[i][w]` 다시 저장한다.
- `return A[n][w]` A[i][w]결과를 반환 


3. 손으로 푼 예제 (이미지삽입):



4. 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def knapSack_dp(W, wt, val, n):` 

- 입력 변수 :  
    - `W` 현재 배낭의 용량
    - `wt` 물건들의 무게를 저장한 리스트
    - `val` 물건들의 가치를 저장한 리스트 
    - `n` 물건들의 수 
- 출력 변수 : 
    - `A[i][w]` 항목을 넣는 경우의 가치합 과 넣지 않는 경우의 가치합 중에서 큰 값


```python
# 5. 코드
def knapSack_dp(W, wt, val, n):
    A = [[ 0 for x in range (W + 1 )] for x in range ( n + 1 )]

    for i in range(1, n+1):
        for w in range(1, W+1):
            if wt[i-1] > w:
                A[i][w] = A[i-1][w]
            else :
                valWith = val[i-1] + A[i-1][w-wt[i-1]]
                valWithout = A[i-1][w]
                A[i][w] = max(valWith, valWithout)

    return A[n][w]
```

```python
# 6. 테스트 코드 
import random
import time

random.seed(0)  
val = [random.randint(1, 100) for _ in range(50)] # 물건들의 가치 리스트 
wt = [random.randint(1, 10) for _ in range(50)] # 물건들의 무게 리스트
W = 25 # 현재 배낭의 용량 
n = len(val)

start = time.time()
result = knapSack_dp(W, wt, val, n)
end = time.time()

print("0-1 배낭문제(동적계획):", result)
print("실행 시간: %.5f 초" % (end - start))
```
